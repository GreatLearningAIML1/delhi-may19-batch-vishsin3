```python
Y = bank_data['Exited']
X = bank_data.drop(['Exited'], axis = 1)
```

# 4. Divide the data set into train and test sets

```python
# Splitting the dataframe into training and test set in the ratio of 70:30 

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 51)
```

# 5. Normalize the train and test data

```python
# Normalizing the data

from sklearn.preprocessing import Normalizer
Normalize = Normalizer()
X_train_norm =  Normalize.fit_transform(X_train) 
X_test_norm =  Normalize.fit_transform(X_test)
```

```python
Y_train = Y_train.astype('float32')
Y_train_array = Y_train.values

Y_test = Y_test.astype('float32')
Y_test_array = Y_test.values
```

# 6. Initialize & build the model

```python
from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout, Flatten, Reshape
```

```python
# Define Model
bank_model = Sequential()

#Adding 1st dense layer
bank_model.add(Dense(200, activation='relu', input_shape=(12,)))

#Add 2nd hidden layer
bank_model.add(Dense(100, activation='relu'))

#Adding OUTPUT layer
bank_model.add(Dense(1, activation='sigmoid'))
```

# 7. Optimize the model

```python
#Compile the model
bank_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
bank_model.fit(X_train_norm, Y_train_array, epochs=10, validation_data=(X_test_norm, Y_test_array))
```

# 8. Predict the results using 0.5 as a threshold

```python
Y_pred = bank_model.predict(X_test_norm)
```

```python
Y_pred = (Y_pred>0.5)
Y_pred
```

# 9. Print the Accuracy score and confusion matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
Matrix = confusion_matrix(Y_test_array, Y_pred)
Acc_score = accuracy_score(Y_test_array, Y_pred)
```
